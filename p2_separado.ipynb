```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Smogon_agrupados.csv")
```

```python
data
```

```python
# Descartar la columna del cluster
data = data.drop("Tipo", axis=1)
```

```python
data
```

```python
#borrar el indice
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
pca=PCA(n_components=10)
pca.fit(data)
x_pca = pca.transform(data)
```

```python
print("Número de filas y columnas del DataFrame original:")
print(data.shape)
```

```python
# Imprimir el número de filas y columnas del DataFrame de componentes principales
print("Número de filas y columnas del DataFrame de componentes principales:")
print(x_pca.shape)
```

```python
cabeceras = ["PCA1", "PCA2", "PCA3", "PCA4", "PCA5", "PCA6", "PCA7", "PCA8", "PCA9", "PCA10"]
```

```python
tablaPCA = pd.DataFrame(data=x_pca, columns=cabeceras)
print(tablaPCA)
```

```python
km = KMeans(n_clusters=18, n_init=40)
clusters_lista = km.fit_predict(tablaPCA)
print(clusters_lista)
```

```python
tablaPCA["cluster"] = clusters_lista
tablaPCA
```

```python
# Guardar el DataFrame con la matriz PCA y el cluster en un archivo CSV
tablaPCA.to_csv("resultado_pca_cluster.csv", index=False)

```
